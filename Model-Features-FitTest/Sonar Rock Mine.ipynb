# **Preprocess**

```python
from sklearn.preprocessing import PowerTransformer,MinMaxScaler,QuantileTransformer,MaxAbsScaler,Normalizer
```

```python
def power_Tranformer_yeo(xtr,xte):
  x_train = PowerTransformer(method='yeo-johnson', standardize=False).fit_transform(xtr);
  x_test = PowerTransformer(method='yeo-johnson', standardize=False).fit_transform(xte);
  x_train = pd.DataFrame(x_train);
  x_test = pd.DataFrame(x_test);
  return x_train,x_test

def minmax(xtr,xte):
  x_train = MinMaxScaler().fit_transform(xtr);
  x_test = MinMaxScaler().fit_transform(xte);
  x_train = pd.DataFrame(x_train);
  x_test = pd.DataFrame(x_test);
  return x_train,x_test

def Quantile(xtr,xte):
  x_train = QuantileTransformer().fit_transform(xtr);
  x_test = QuantileTransformer().fit_transform(xte);
  x_train = pd.DataFrame(x_train);
  x_test = pd.DataFrame(x_test);
  return x_train,x_test

def MaxAbs(xtr,xte):
  x_train = MaxAbsScaler().fit_transform(xtr);
  x_test = MaxAbsScaler().fit_transform(xte);
  x_train = pd.DataFrame(x_train);
  x_test = pd.DataFrame(x_test);
  return x_train,x_test

def Normalizer_pre(xtr,xte):
  x_train = Normalizer().fit(xtr).transform(xtr);
  x_test = Normalizer().fit(xte).transform(xte);
  x_train = pd.DataFrame(x_train);
  x_test = pd.DataFrame(x_test);
  return x_train,x_test
```

```python
# Import Other Tools
from sklearn.model_selection import train_test_split,RepeatedStratifiedKFold,cross_val_score
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,f1_score
from sklearn.pipeline import Pipeline
from sklearn import preprocessing
from numpy import mean

# Import Model
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# Import Feature Selection
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.linear_model import Lasso
from sklearn.decomposition import FastICA
from sklearn.manifold import LocallyLinearEmbedding
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
# from sklearn.feature_selection import RFE

# Simple value
model_list = ['Random Forest','Gradient Boosting','Logistic Regression','Decision Tree','GaussianNB','Support_Vector_Machines(SVM)','K-Nearest_Neighbors(KNN)']
model = [RandomForestClassifier(),GradientBoostingClassifier(),LogisticRegression(),DecisionTreeClassifier(),GaussianNB(),SVC(),KNeighborsClassifier()]
stay = 0
chek += 1
f_s = [0,0,0,0,0,0,0]
s_f = ['t-SNE','PCA','LLE','ICA','PCA_t-SNE','LLE_t-SNE','ICA_t-SNE']
```

```python
# Simeple Functions
def staybility_test(model,x_train,y_train,x_test,y_test,n):
  rou = []
  model_fit = []
  for i in range(n):
    for j in range(len(model)):
      model_fit.append(model[j].fit(x_train,y_train));
      rou.append(model_fit[j].score(x_test,y_test));
  rou = np.array(rou).reshape(n,len(model));
  return rou

def classifier(x_train,x_test,y_train,y_test,Name_of_feature,n_of_staybility_test,frombast=False):
  
  # Create model and Fit model
  model_fit = []
  pred = []
  for i in range(len(model_list)):
    model_fit.append(model[i].fit(x_train,y_train));
    pred.append(model_fit[i].predict(x_test));

  # Prediction
  # for i in range(len(model_fit)):
  #   pred.append(model_fit[i].predict(x_test))

  if n_of_staybility_test != 0 :
    rou = staybility_test(model,x_train,y_train,x_test,y_test,n_of_staybility_test);

  # Show result
  score_ = []
  if frombast == False:
    for i in range(len(pred)):
      m = 0;
      print(model_list[i],':',round(model[i].score(x_test,y_test),5));
      score_.append(model[i].score(x_test,y_test));
      print('Staybility_test');
      for j in range(n_of_staybility_test):
        print('round',j+1,' :',rou[j][i]);
        m += rou[j][i];
      if n_of_staybility_test != 0:
        print('Mean :',m/n_of_staybility_test);
      print(confusion_matrix(y_test,pred[i]));
      print(classification_report(y_test,pred[i]));
      print(li);

  #keep in to result
  if Name_of_feature != 0:
    result[Name_of_feature] = [model[i].score(x_test,y_test) for i in range(len(model))];

  return score_

# Feature selection then split
def bast_decition_f_s(chek):
  if chek == 0:
    x_train,x_test,y_train,y_test = train_test_split(origin_x,origin_y,test_size=0.3);
    classifier(x_train,x_test,y_train,y_test,'standard',0,frombast=True);
    
  #t-SNE
  x_tsne = TSNE(n_components=3).fit_transform(origin_x);
  x_train,x_test,y_train,y_test = train_test_split(x_tsne,origin_y,test_size=0.3);
  f_tsne = classifier(x_train,x_test,y_train,y_test,'t-SNE',0,frombast=True);

  #PCA
  pca = PCA(n_components=3).fit_transform(origin_x);
  x_train,x_test,y_train,y_test = train_test_split(pca,origin_y,test_size=0.3);
  f_pca = classifier(x_train,x_test,y_train,y_test,'PCA',0,frombast=True);
  
  #LLE
  lle = LocallyLinearEmbedding(n_components=3,eigen_solver ="dense").fit_transform(origin_x);
  x_train,x_test,y_train,y_test = train_test_split(lle,origin_y,test_size=0.3);
  f_lle = classifier(x_train,x_test,y_train,y_test,'LLE',0,frombast=True);

  #ICA
  ica = FastICA(n_components=3).fit_transform(origin_x);
  x_train,x_test,y_train,y_test = train_test_split(ica,origin_y,test_size=0.3);
  f_ica = classifier(x_train,x_test,y_train,y_test,'ICA',0,frombast=True);
  
  #PCA + t-SNE
  pca = PCA(n_components=6).fit_transform(origin_x);
  pca_tsne = TSNE(n_components=3).fit_transform(pca);
  x_train,x_test,y_train,y_test = train_test_split(pca_tsne,origin_y,test_size=0.3);
  f_pcatsne = classifier(x_train,x_test,y_train,y_test,'PCA_t-SNE',0,frombast=True);
  
  #LLE + t-SNE
  lle = LocallyLinearEmbedding(n_components=6,eigen_solver ="dense").fit_transform(origin_x);
  lle_tsne = TSNE(n_components=3).fit_transform(lle);
  x_train,x_test,y_train,y_test = train_test_split(lle_tsne,origin_y,test_size=0.3);
  classifier(x_train,x_test,y_train,y_test,'LLE_t-SNE',0,frombast=True);

  #ICA + t-SNE
  ica = FastICA(n_components=6).fit_transform(origin_x);
  ica_tsne = TSNE(n_components=3).fit_transform(ica);
  x_train,x_test,y_train,y_test = train_test_split(ica_tsne,origin_y,test_size=0.3);
  f_ica_tsne = classifier(x_train,x_test,y_train,y_test,'ICA_t-SNE',frombast=True);

  chek += 1
  return chek

# Split then feature selection

def bast_decition_s_f(chek,pre = False):
  x_train,x_test,y_train,y_test = train_test_split(origin_x,origin_y,test_size=0.3);
  if pre == False:
    pass
  elif pre == 'power':
    x_train,x_test=power_Tranformer_yeo(x_train,x_test);
    print('pewer')
  elif pre == 'minmax':
    x_train,x_test=minmax(x_train,x_test);
    print('minmax')
  elif pre == 'Quantile':
    x_train,x_test=Quantile(x_train,x_test);
    print('Quantile')
  elif pre == 'MaxAbs':
    x_train,x_test=Quantile(x_train,x_test);
    print('MaxAbs')
  elif pre == 'Normalizer_pre':
    x_train,x_test=Normalizer_pre(x_train,x_test);
    print('Normalizer_pre')

  if chek == 0:
    classifier(x_train,x_test,y_train,y_test,'standard',0,frombast=True);

  #t-SNE
  x_train_tsne = TSNE(n_components=3).fit_transform(x_train);
  x_test_tsne = TSNE(n_components=3).fit_transform(x_test);
  s_tsne = classifier(x_train_tsne,x_test_tsne,y_train,y_test,'t-SNE',0,frombast=True);

  #PCA
  x_train_pca = PCA(n_components=3).fit_transform(x_train);
  x_test_pca = PCA(n_components=3).fit_transform(x_test);
  s_pca = classifier(x_train_pca,x_test_pca,y_train,y_test,'PCA',0,frombast=True);

  #LLE
  x_train_lle = LocallyLinearEmbedding(n_components=3,eigen_solver ="dense").fit_transform(x_train);
  x_test_lle = LocallyLinearEmbedding(n_components=3,eigen_solver ="dense").fit_transform(x_test);
  s_lle = classifier(x_train_lle,x_test_lle,y_train,y_test,'LLE',0,frombast=True);
  
  #ICA
  x_train_ica = FastICA(n_components=3).fit_transform(x_train);
  x_test_ica = FastICA(n_components=3).fit_transform(x_test);
  s_ica = classifier(x_train_ica,x_test_ica,y_train,y_test,'ICA',0,frombast=True);
  
  #PCA + t-SNE
  x_train_pca = PCA(n_components=6).fit_transform(x_train);
  x_train_pca_tsne = TSNE(n_components=3).fit_transform(x_train_pca);
  x_test_pca = PCA(n_components=6).fit_transform(x_test);
  x_test_pca_tsne = TSNE(n_components=3).fit_transform(x_test_pca);
  s_pcatsne = classifier(x_train_pca_tsne,x_test_pca_tsne,y_train,y_test,'PCA_t-SNE',0,frombast=True);

  #LLE + t-SNE
  x_train_lle = LocallyLinearEmbedding(n_components=6,eigen_solver ="dense").fit_transform(x_train);
  x_train_lle_tsne = TSNE(n_components=3).fit_transform(x_train_lle);
  x_test_lle = LocallyLinearEmbedding(n_components=6,eigen_solver ="dense").fit_transform(x_test);
  x_test_lle_tsne = TSNE(n_components=3).fit_transform(x_test_lle);
  s_lletsne = classifier(x_train_lle_tsne,x_test_lle_tsne,y_train,y_test,'LLE_t-SNE',0,frombast=True);

  #ICA + t-SNE
  x_train_ica = FastICA(n_components=6).fit_transform(x_train);
  x_train_ica_tsne = TSNE(n_components=3).fit_transform(x_train_ica);
  x_test_ica = FastICA(n_components=6).fit_transform(x_test);
  x_test_ica_tsne = TSNE(n_components=3).fit_transform(x_test_ica);
  s_ica_tsne = classifier(x_train_ica_tsne,x_test_ica_tsne,y_train,y_test,'ICA_t-SNE',0,frombast=True);
  
  chek += 1
  return chek

# Build score bord
def score_m_f(result,result_point,chek,chek1):
  if chek1 == 0:
    result_point[[result.iloc[:,1:].columns]] = 0;
  if chek > chek1:
    chek1 = chek;
    for i in model_list:
      for j in range(result.shape[1]-1):
        if pd.notna(result.iloc[0,1:].round(4).where(result.loc[i]==result.iloc[:,1:].loc[i].max())).values[j] == True:
          result_point.loc[i][j] += 1;
  print('Round :',chek)
  return result_point,chek1
```
