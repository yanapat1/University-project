# **Load data**

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.model_selection import train_test_split
#Amount sample
n=2000
```

### Train set data

```python
#Train animal
load_train_cat = '/content/gdrive/MyDrive/share/afhq/train/cat/'
load_train_dog = '/content/gdrive/MyDrive/share/afhq/train/dog/'
load_train_wild = '/content/gdrive/MyDrive/share/afhq/train/wild/'

#Read images and changed it to gray scale
#and keep it in list
def load_img_data(path):
  lis = []
  for filename in os.listdir(path):
      img = cv2.imread(os.path.join(path,filename))
      if img is not None:
          img1 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
          lis.append(img1)
  print("Finish load path:",path)
  return lis

train_cat = load_img_data(load_train_cat)
train_dog = load_img_data(load_train_dog)
train_wild = load_img_data(load_train_wild)
```
